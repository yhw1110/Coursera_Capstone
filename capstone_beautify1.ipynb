# Applied data science capstone: Segmenting and Clustering Neighborhoods in Toronto

```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request
import requests
import numpy as np
np.nan
```

### Scape table from wikipedia using requests and beautifulsoup

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
r1= requests.get(url)
html_content =r1.content
print(html_content)
```

```python
soup= BeautifulSoup(html_content,"html.parser")
stat_table=soup.find_all('table',class_="wikitable sortable")
len(stat_table)
```

```python
table1=stat_table[0]
```

```python
for j,row in enumerate(table1.find_all('tr')):
    for k, cell in enumerate(row.find_all('td')):  
        if k==2:
            print(cell.text, end=' ')
        else:
            print(cell.text, end=',')    
    if j==20:
        break
       
```

```python
with open('wiki_table.csv','w') as file1:
    for row in table1.find_all('tr'):
        for k, cell in enumerate(row.find_all('td')):  
            if k==2:
                file1.write(cell.text)
            else:
                file1.write(cell.text)
                file1.write(",")

path='wiki_table.csv'
dft=pd.read_csv(path, header=None)
dft.columns=['PostalCode','Borough','Neighborhood']
dft.head(10)
```

### Replace "not assigned" by NaN

```python
dft.replace("Not assigned",np.nan,inplace=True)
dft.dropna(subset=["Borough"],axis=0,inplace=True)
```

```python
dft[0:10]
```

###  Replace NaN in the Neighborhood column by the value from the Borough column

```python
for p in dft.index:
    if  dft.loc[p,"Neighborhood"] is np.nan:
        dft.loc[p,"Neighborhood"]=dft.loc[8,"Borough"]
```

```python
dft
```

### Grouping the data frame: merge the first two columns, combine the entries for the third column, separated by ","

```python
dfg=pd.DataFrame() 
dfg['Neighborhood']=dft.groupby(['PostalCode','Borough'])['Neighborhood']\
.apply(lambda x: "%s" % ', '.join(x))
```

```python
dfg.reset_index(level=['PostalCode','Borough'],inplace=True)
```

```python
print("(row number,column number)=",dfg.shape)
```

```python
df_cor=pd.read_csv('http://cocl.us/Geospatial_data')
df_cor.set_index('Postal Code',inplace=True)
df_cor.head()
```

```python
dfg2=dfg.set_index('PostalCode')
dfg_index=dfg2.index
df_cor=df_cor.loc[dfg_index]
df_cor = df_cor.reset_index(drop=True)
```
